```python
#Loading city bike data for April-October 2018
bikes04_18 = pd.read_csv('2018-07.csv')
bikes05_18 = pd.read_csv('2018-07.csv')
bikes06_18 = pd.read_csv('2018-07.csv')
bikes07_18 = pd.read_csv('2018-07.csv')
bikes08_18 = pd.read_csv('2018-08.csv')
bikes09_18 = pd.read_csv('2018-09.csv')
bikes10_18 = pd.read_csv('2018-10.csv')
```
